```python
from bs4 import BeautifulSoup, SoupStrainer
import requests
from urllib.request import urlopen
import pandas as pd
import numpy as np
```

```python
url = 'https://www.cs.cmu.edu/~mleone/gdead/95.html'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

```python

url2 = 'https://www.cs.cmu.edu/~mleone/gdead/dead-sets/95/2-19-95.txt'

with urlopen(url2) as link:
    result = link.read()

soup2 = BeautifulSoup(result)

setlist_raw = soup2.find('p')
setlist = setlist_raw.text

print(setlist_raw)
```

```python
data = setlist.split('\n')
```

```python
setlist_df = pd.DataFrame(data)
setlist_df[0].replace('', np.nan, inplace=True)
setlist_df.dropna(subset=[0], inplace=True)
setlist_df[0] = setlist_df[0].str.replace('->','')
setlist_df
```

```python
concert = setlist_df.iloc[0,0]
print(concert)
```

```python
setlist_df.drop([0], axis=0, inplace=True)
setlist_df['Concert'] = concert
setlist_df['Year'] = '1995'
setlist_df.rename({0:'Song'}, axis=1, inplace=True)
setlist_df
```

```python
print(data)
```

```python
concert_details = data.pop(0)
print(concert_details)

songs = list(data)
print(songs)
```

```python
date_str = concert_details.split()[-1:]
date = date_str[0]
year_2dig = date[-3:]
year_2dig = year_2dig.replace(')', '')
print(year_2dig)
```

```python
concert_all = {'concert_details':concert_details, 'setlist':songs, 'Year':'19'+year_2dig}
```
